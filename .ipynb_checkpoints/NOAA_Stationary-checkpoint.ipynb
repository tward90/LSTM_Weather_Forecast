```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
NOAA_df = pd.read_csv('NOAA_Data.csv', na_values=[9999.9, 999.9, 99.99])
NOAA_df.info()
```

```python
df_drop = NOAA_df.drop(["stp", "gust", "tornado_funnel_cloud", "sndp"], axis=1)
df_drop = df_drop.dropna()
df_drop.info()
```

```python
DateTimeConvert = pd.to_datetime(df_drop.year*10000+df_drop.mo*100+df_drop.da, format='%Y%m%d')
```

```python
df_drop['Date'] = DateTimeConvert
```

```python
df_drop.columns
```

```python
NOAA_trim = df_drop.drop(['Unnamed: 0', 'year', 'mo', 'da'], axis=1)
NOAA_trim.head()
```

```python
NOAA_trim = NOAA_trim.set_index('Date')
NOAA_trim.head()
```

```python
NOAA_sorted = NOAA_trim.sort_index()
NOAA_sorted.info()
```

```python
# data = pd.read_csv('NOAA_Data.csv')
values = NOAA_sorted.values
groups = np.arange(14)
i=1
plt.figure(figsize=(15,10))
for group in groups:
    plt.subplot(len(groups), 1,i)
    plt.plot(values[:,group])
    plt.title(NOAA_sorted.columns[group], y=0.5, loc='right')
    i += 1
plt.show()
```
